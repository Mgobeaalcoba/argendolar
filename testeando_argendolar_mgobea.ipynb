```python
# @title Método que me devuelve la totalidad de las cotizaciones de un tipo de dolar solicitado desde que existe registro informático
historical_data = argendolar.get_dolar_historia_completa(TipoDivisas.MEP)
historical_data.tail()
```

```python
# @title Uso librerías para mostrar la evolución del MEP que traje de mi clase Argendolar by Mgobea
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# @title Gráfico en serie temporal la evolución del MEP
# historical_data['fecha'] = pd.to_datetime(historical_data['fecha']) # Descomentar la primera vez

# historical_data.set_index('fecha', inplace=True) # Descomentar la primera vez

plt.figure(figsize=(10, 6))
plt.plot(historical_data.index, historical_data['compra'], label='Compra')
plt.plot(historical_data.index, historical_data['venta'], label='Venta')

plt.title('Evolución del Dólar MEP a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Tipo de Cambio')
plt.legend()

plt.show()
```
